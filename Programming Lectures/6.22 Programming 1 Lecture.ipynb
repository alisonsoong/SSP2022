```python
# 4
from math import *
def solveQuad(a, b, c):
    det = b*b - 4*a*c
    if det==0: return [-b/(2*a), '+/-', 0]
    elif det>0: return [-b/(2*a), '+/-', sqrt(det)/(2*a)]
    else: return [-b/(2*a), '+/- i', sqrt(abs(det))/(2*a)]
    
solveQuad(1,4,4)
```

```python
# 5
import math
def leibnitz(n):
    res = 0
    for i in range(n): res+=((-1)**i)*1/(i*2+1)
    res*=4
    print("Approximation: ", res, " ; Error: ", math.pi-res)

leibnitz(100)
    
```

```python
# 6
from random import randint

def randomGame():
    num=randint(1,100)
    while True:
        try:
            g = int(input("Guess: "))
            if g<num: print("Too low")
            elif g>num: print("Too high")
            else: break
        except: print("Enter a valid integer.")
                
    print("you guessed it :D")
    
randomGame()
```

```python
# 7
def getPrimeN(n):
    res = []
    
    for i in range(2,n+1):
        flag = False
        for j in range(2,i): 
            if i%j==0: 
                flag = True
                break
        if not flag: res.append(i)
    
    return [j for j in [(i,j) for i in range(2,n+1) for j in range(2,i) if i%j == 0]
    

getPrimeN(100)
```
