```python
import numpy as np
#1
def populate(n):
    if (n>10): raise Exception("Value >10 is not accepted")
    arr = np.zeros((n,n))
    for i in range(n):
        arr[i, 0:n] = np.arange(10*i,10*i+n,1)
    return arr.astype(int)

populate(6)
```

```python
#2
import numpy as np
def det3x3(m):
    return m[0,0]*(m[1,1]*m[2,2] - m[1,2]*m[2,1]) - m[0,1]*(m[1,0]*m[2,2] - m[1,2]*m[2,0]) + m[0,2]*(m[1,0]*m[2,1] - m[1,1]*m[2,0])

det3x3(np.array([[1,2,3],[3,2,1],[5,4,3]]))
```

```python
#3 
import matplotlib.pyplot as plt
import numpy as np

def main3(A, sigma):
    X = np.arange(-5,5,0.1)
    Y1 = np.sin(X)
    Y2 = np.cos(X)
    Y3 = np.sin(X) * np.cos(X)
    Y4 = A * np.exp(-np.power(X,2)/sigma)
    plt.plot(X,Y3)
    plt.plot(X,Y2)
    plt.plot(X,Y1)
    
    plt.figure()
    plt.plot(X,Y4)
    
    plt.show()
    
main3(5,4)
```

```python
#4
import numpy as np
from math import *

def main4(A,sigma,res):
    result = np.zeros((res,res))
    for x in range(res):
        for y in range(res):
            result[x,y]=round(A*pow(e,-(pow(x-int(res/2),2)+pow(y-int(res/2),2))/sigma),1)
    
    plt.imshow(result)
    return result
    
main4(20,4,40)
```
