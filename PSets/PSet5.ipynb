```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#5 
def newton(func,der,init,err): #function, derivative, inital_guess, error_tolerance
    prev=-1
    while abs(init-prev)>=err:
        prev,init=init,init-func(init)/der(init)
    return init

print("E =", newton(lambda x:3*math.pi/4 - x - 0.41*np.sin(x), lambda x: -1-0.41*np.cos(x), 3*math.pi/4,1e-10 ))
```

```python
#11
def prob11(x, N, rng, graph=False):
    X=np.arange(0,x+rng,0.01)
    Y=np.sin(X)
    approx=np.array([sum([val**(2*i+1)/(math.factorial(2*i+1))*(-1)**i for i in range(N)]) for val in X])
    estimate=sum([x**(2*i+1)/(math.factorial(2*i+1))*(-1)**i for i in range(N)])
    actual=math.sin(x)
    
    if graph:
        plt.plot(X,Y,'ro')
        plt.plot(X,approx,'bo',markersize=2)
        plt.text(1.4,-0.5,"N: "+str(N)+"\nx: "+str(x))
        plt.ylim(-2,2)
        plt.show()
        print("x:",x,"\nN:",N)
        print("estimate",estimate,"\nactual",actual,"\nerror",estimate-actual)
    return estimate, actual
    
prob11(3,5,5,True)
prob11(3,20,15,True)
prob11(3,50,40,True) # limit for N

print()
def check():
    prev=0
    for val in np.arange(1,6,0.3):
        for i in range(30,120):
            try:
                est,act = prob11(val,i)
            except: break
            #print(est,act)
            if round(est,16)==round(prev,16): 
                print("DONE: N =",i,"val =",val)
                break
            else: 
                prev=est 
```
