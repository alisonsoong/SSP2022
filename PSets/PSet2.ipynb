```python
# 4
def DefInt(eq:str, a:float, b:float, N:int)->float:
    '''Solves a definite integral with (a,b) inclusive as the limits of integration, 
        as a midpoint Riemann sum w/ N intervals'''
    interval = float((b-a))/N
    X = np.arange(a+interval/2,b,interval)
    Y = eval(eq)
    return np.sum(Y)*interval

n = 1
while True:
    if abs(DefInt("np.sqrt(np.power(X,3)+2*np.sin(10/X))",4,12,n) - 187.11) < 0.005:
        print(n)
        break
    n+=1

# 187.10 is the answer to 5 significant figures

def ActDefInt(eq, a, b, N):
    '''Solves a definite integral with (a,b) inclusive as the limits of integration, 
        as a midpoint Riemann sum w/ N intervals'''
    interval = float((b-a))/N
    X = np.arange(a+interval/2,b,interval)
    return sum([eq(x) for x in X])*interval

def Troll(eq,a:float,b:float,N:float)->float:
    return sum([eq(x) for x in np.arange(a+float((b-a))/N/2,b,float((b-a))/N)])*float((b-a))/N

Troll(lambda x:sin(x), 0, 5, 100)
```

```python
# 5
def GetAngle(sin:float,cos:float)->float:
    '''Returns the arcsin of a value as an angle (in radians) in the correct quadrant'''
    v = math.atan2(cos,sin)
    if cos>0 and sin>0: # first quadrant
        return v
    elif cos<0 and sin>0: # second quadrant
        return v+pi
    elif cos<0 and sin<0: # third quadrant
        return v+2*pi
    elif cos>0 and sin<0: # fourth quadrant
        return v+pi
    elif cos==0:
        if sin==1: return pi/2
        if sin==-1: return 3*pi/2
    elif sin==0:
        if cos==1: return 0
        if cos==-1: return pi
    else: return -1

GetAngle(1/np.sqrt(2), -1/np.sqrt(2))
```

```python
# 6a
def DMStoDeg(d:float,arcm:float,arcs:float)->float:
    return math.copysign(abs(d)+arcm/60+arcs/3600,d)

DMStoDeg(-30,30,30)
```

```python
# 6b
def RAdecimalToHMS(dec:float)->tuple:
    dec%=360
    h=dec/15
    m=(dec%15)/15*60
    s=m%1*60
    h//=1
    m//=1
    print((h,m,s))
    return (h,m,s)

RAdecimalToHMS(56.3)
```

```python
# 6d (where is c)
def DECdecimalToDMS(dec:float)->tuple:
    d=math.trunc(dec)
    dec=abs(dec)-abs(d)
    
    m=math.trunc(dec*60)
    dec-=m/60
    s=dec*3600
    if (round(s,0)-s<0.00001):s=round(s,0)
    print((d,m,s))
    return (d,m,s)

DECdecimalToDMS(-125.32)
```

```python
# 6e
def magnitude(v)->float:
    return np.sqrt(v[0]*v[0]+v[1]*v[1]+v[2]*v[2])

magnitude(np.array([-1, -2, 3]))
```

```python
# 7c
def rot(v,alpha, beta):
    z=np.array([[np.cos(alpha),np.sin(alpha),0],
                [-np.sin(alpha),np.cos(alpha),0],
                [0,0,1]])
    x=np.array([[1,0,0],
                [0,np.cos(beta),np.sin(beta)],
                [0,-np.sin(beta),np.cos(beta)]])
    return np.dot(np.matmul(x,z),v)

rot(np.array([1, 1, 1]), np.pi/2, 0)
```

```python
# 8a
def spiral():
    theta=np.arange(0,50,0.1)
    X=np.sqrt(theta)*np.cos(theta)
    Y=np.sqrt(theta)*np.sin(theta)
    plt.plot(X,Y)
    plt.axis("square")
    
spiral()
```
